## the primary goal of this pipeline is to perform standard data preprocessing as in AFNI proc.py, then perform subject-level GLM analysis against the HRF-convolved event paradigm, correct for temporal autocorrelation with subsequent REML analysis, then threshold and cluster resultant coefficients to get subject-level 3d brain masks
## in addition, this pipeline will extract time series data from the subject-level brain masks and/or predetermined ROI brain masks

## Helper function to easily index CBV versus func scans when defining group filters and for CBV correction later

```python
#fill in your folder names
def gorg_helper(pipe):
    func = None
    mion = None
    for k, v in pipe.executed.items():
        if 'last-func-preprocessing-step' in v:
            func = k
        if 'last-cbv-preprocessing-step' in v:
            mion = k
    if any([(a == None) for a in [func]]):
        raise Exception
    elif mion is None:
        return dict(func=func)
    else:
        return dict(func=func, mion=mion)
```

```python
indices = gorg_helper(pipe)
print(indices)
```

```python
# Use this to generate a one-sample ttest 3d brain map from subject-level coefficients from the GLM and REML analyses in the target group
# This will also generate a clustsim file that can be used to threshold group-level maps based on FWE correction for multiple comparisons
import os
cmd = '3dttest++ -mask path-to-whole-or-partial-brain-mask -prefix -path-to-output -resid -path-for-residuals -ACF -prefix_clustsim -path-for-FWE-clustim-result -clustsim -setA '
cmd = cmd + ' '.join([os.path.relpath(f)+'"[1]"' for f in prj(1, 'sub-project-folder-name', 'subject-coefficients-folder-name', 'group-name',).df.Abspath])
print(cmd)
out, err = pn.shell(cmd)
print(out, err)
```

```python
# use this to generate a 3d brain map of subject-level super-threshold cluster masks (each subject = 1, map shows sum)
cmd = '3dMean -prefix output-path-and-file-name -count '
cmd = cmd + ' '.join([f for f in prj(1, 'subproject-folder-name', 'subject-clusters-folder-name', 'group-name', ignore=[
    'subjects-to-ignore']).df.Abspath])
print(cmd)
out, err = pn.shell(cmd)
print(out, err)
```

```python
# use this to generate a two-sample ttest 3d brain map for subject-level coefficients in one group versus another
cmd = '3dttest++ '
cmd1 = cmd + '-setA '+' ' .join([f+'"[1]"' for f in prj(1, 'subproject-folder-name', 'subject-coefficient-folder-name', 'group_one-name', ).df.Abspath])
cmd2 = cmd1 + ' -setB '+' '.join([g+'"[1]"' for g in prj(1, 'subproject-folder-name', 'subject-coefficient-folder-name', 'group_two-name', ).df.Abspath])
cmd3 = cmd2 + ' -prefix output-path-and-file-name -mask "directory-to-brain-mask-if-applicable" -paired'
print(cmd3)
out, err = pn.shell(cmd3)
print(out, err)
```
